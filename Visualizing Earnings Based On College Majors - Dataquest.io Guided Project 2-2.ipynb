### Some Notes on Observations
Based on the data presented in the graphs, it does not clearly show that more popular majors make more money.

Additionally, it is not clear that majors with majority females make more money.

As for a link between full-time employees and median income, there seems to be a very loose negative correlation in that the more full time positions in a major, the less the income.
